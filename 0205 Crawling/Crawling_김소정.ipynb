```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

path = 'C:/Users/sojeo/Downloads/chromedriver_win32/chromedriver.exe'
driver = webdriver.Chrome(path)

import re
import requests
```

```python
cine21_url = 'http://www.cine21.com/rank/person/content'
post_data = dict()
actors_info_list = list()

for i in range(0, 11):
    post_data['section'] = 'actor'
    post_data['period_start'] = '2022-01'
    post_data['gender'] = 'all'
    post_data['page'] = i

    res = requests.post(cine21_url, data=post_data)
    
    soup = BeautifulSoup(res.text, 'html.parser')
    
    soup.select('li.people_li div.name')
    
    actors = soup.select('li.people_li div.name')
    
    grades = soup.select('li.people_li span.grade')
    grades_list = []
    for grade in grades:
        grades_list.append(grade.text)
        
    indices = soup.select('li.people_li ul.num_info li strong')
    indices_list = []
    for index in indices:
        indices_list.append(index.text)
        
    movies = soup.select('span')
    mov_list = []
    for movie in movies:
        mov = movie.get_text()
        mov_list.append(mov)
    
    for index, actor in enumerate(actors):
        actor_info_dict = dict()
        actor_name = re.sub('\(\w*\)', '', actor.text)
        actor_info_dict['이름'] = actor_name
        actor_info_dict['랭킹'] = grades_list[index]
        actor_info_dict['흥행지수'] = indices_list[index]
        
        firstidx = mov_list.index('흥행지수')
        lastidx = mov_list.index(grades_list[index])
        movie_list = mov_list[firstidx + 1:lastidx]

        mov_list = mov_list[lastidx + 1:]
        
        actor_info_dict['출연영화'] = movie_list
        
        actor_link = 'http://www.cine21.com' + actor.select_one('a')['href']
        response_actor = requests.get(actor_link)
        soup_actor = BeautifulSoup(response_actor.text, 'html.parser')
        default_info = soup_actor.select_one('ul.default_info')
        actor_details = default_info.select('li')
        
        for actor_detail in actor_details:
            actor_detail_key = actor_detail.select_one('span.tit').text
            actor_detail_value = re.sub('<span.*>.*</span>', '', str(actor_detail))
            actor_detail_value = re.sub('<.*?>', '', actor_detail_value)
            actor_info_dict[actor_detail_key] = actor_detail_value
        
        actors_info_list.append(actor_info_dict)
        
actors_info_list
```
